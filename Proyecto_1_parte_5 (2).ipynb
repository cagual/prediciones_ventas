```python
df = df.astype({"Item_Outlet_Sales":'int'})
```

```python
X = df[['Item_MRP', 'Outlet_Location_Type']].values
y = df['Item_Outlet_Sales'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=20,test_size=0.3,train_size=0.7)
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, random_state=500,test_size=0.5,train_size=0.5)

scaler = MinMaxScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test = scaler.transform(X_test)
X_val = scaler.transform(X_val)
```

```python
y_test
```

#KNN VECINOS

```python
neighbors = np.arange(1, 35)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

# Loop over K values
for i, k in enumerate(neighbors):
	knn = KNeighborsRegressor(n_neighbors=k)
	knn.fit(X_train, y_train)
	# Compute training and test data accuracy
	train_accuracy[i] = knn.score(X_train, y_train)
	test_accuracy[i] = knn.score(X_test, y_test)

print(train_accuracy)
print(test_accuracy)
# Generate plot
plt.plot(neighbors, test_accuracy, label = 'Testing dataset Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training dataset Accuracy')

plt.legend()
plt.xlabel('n_neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
n_neighbors = 10 #me quedo con 10 vecinos porque visualmente es el que el promedio esta mas alto pero un bias entre test y training no tan alto

knn = KNeighborsRegressor(n_neighbors)
knn.fit(X_train, y_train)
print('Accuracy of K-NN Regressor on training set: {:.2f}'
     .format(knn.score(X_train, y_train)))
print('Accuracy of K-NN Regressor on test set: {:.2f}'
     .format(knn.score(X_test, y_test)))
```

```python
y_pred = knn.predict(X_test)
```

```python
# R Square Error
r2 = r2_score(y_test,y_pred)

# Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test,y_pred)

# Mean Square Error (MSE)
mse = mean_squared_error(y_test,y_pred, squared=True)

# Root Mean Square Error (RMSE)
rmse = mean_squared_error(y_test,y_pred, squared=False)

print('R2  :',r2)
print('MAE :',mae)
print('MSE :',mse)
print('RMSE:',rmse)
```
